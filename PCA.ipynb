```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer        #in-built dataset in scikit lib
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()             #30 dimensions which we have to convert into 2D or 3D
```

```python
from sklearn.preprocessing import StandardScaler       #we have to scale our data so that all the features have single unit variance 
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
#PCA
from sklearn.decomposition import PCA                
```

```python
pca = PCA(n_components=2)            #we are gonna visualize the whole dataset df just by using 2 components
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape                  #30 components
```

```python
x_pca.shape                        #2 components->first two principle components
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'], cmap='plasma')    #plotting of all the rows from col 0 to all the rows from col 1
plt.xlabel('First Principle Component')
plt.ylabel('Second Principle Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_comp, cmap='plasma')
```

```python
#PCA is basically the combination of all the features and visualized using heatmap
```
